```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#membuat 3D Axes harus membangun sendiri objek figure dan axesnya
fig = plt.figure()
ax = plt.axes(projection='3d')

plt.show()
```

```python
#membuat plotting sederhana
z = np.linspace(0, 1, 100) #parameter: (nilai terkecil, nilai terbesar, banyak datapoint)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d') #tipe proyeksi

ax.plot(x, y, z, color='red') #parameter ax.plot: (sumbu x, sumbu y, data, color='warna'), parameter color bersifat opsional

ax.set(title='Contoh 3D Line Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')

plt.show()
```

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x, y, z, color='green') #membuat scatter plot. Paramerter: (sumbu x, sumbu y, data, color='warna'), parameter color bersifat opsional

ax.set(title='Contoh 3D Scatter Plot',
       xlabel='Sumbu X',
       ylabel='Sumbu Y',
       zlabel='Sumbu Z')

plt.show()
```

```python
n = np.linspace(-3.0, 3.0, 100) #deret bilangan digenerate dengan np.linspace(nilai terkecil, nilai terbesar, banyak datapoint)

x, y = np.meshgrid(n, n) #membentuk area permukaan terdiri dari x dan y dengan menggunakan np.meshgrid(n, n), memiliki shape/dimensi n*n dari banyak datapoint yang ditentukan
z = np.sqrt(x**2 + y**2) #variabel z berisi hasil akar kuadrat dari sumbu x pangkat 2 ditambah sumbu y pangkat 2

print(x.shape) #sumbu x, y, z berbentuk 2 dimensi
print(y.shape)
print(z.shape)
```

```python
#contour plot dasarnya merupakan visualisasi data yang bisa digunakan untuk mengvisualisasikan data 3D ke 2D. Contour plot biasanya ada di data geospacial.
plt.contour(x, y, z) #data 3D, ditampilkan dalam bentuk 2D
plt.colorbar() #memunculkan bar untuk merepresentasikan warna ke dalam bentuk numerik

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Filled Contour plot memenuhi area dengan warna
plt.contourf(x, y, z)
plt.colorbar()

plt.title('Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
n = np.linspace(-3.0, 3.0, 100)

x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d') #Menampilkan plotting yang bersifat 3D

ax.plot_wireframe(x, y, z, color='orange')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

# ax.plot_surface(x, y, z, color='orange')
ax.plot_surface(x, y, z, cmap='viridis') #cmap adalah parameter untuk membuat plot dengan warna seperti pada contour plot

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Surface Plot')

plt.show()
```
