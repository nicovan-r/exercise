```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([2, 5, 7, 11])
plt.ylabel('sumbu y')
plt.show()
#plt.show untuk menyembunyikan hasil dari jalannya plt.ylabel('sumbu y')
```

```python
plt.plot([2, 3, 4, 7], [5, 4, 2, 16])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'go') #'go' artinya warna hijo:"g" dan "o" bentuk marker/jenis plotting, o bentuk titik.
#"g-" untuk jenis garis, "g*" untuk bentuk bintang
plt.axis([0, 6, 0, 20]) #plt.axis([nilai terkecil x, nilai terkecil x, nilai terkecil y, nilai terbesar max y])
plt.show()
```

```python
t = np.arange(0., 5., 0.2)#np.arange(plotting point terkeci 0, plotting point terbesar 5 angka 5 tidak termasuk, interval konsisten) '.' adalah ploating point
t
#beda dengan linspace yaitu parameter ketiga linspace adalah berapa jumlah data
```

```python
plt.plot(t, t, 'r--') #plt.plot(nilai x, nilai y, warna dan bentuk garis plotting). -- garis putus"
plt.plot(t, t**2, 'bs')#'bs' warna biru dengan bentuk persegi
plt.plot(t, t**3, 'g^')#'g^' warna hijau dengan bentuk segitiga
plt.show()
```

```python
plt.plot(t, t, 'r--', 
         t, t**2, 'bs', 
         t, t**3, 'g^') #memanggil satu kali plt.plot bisa banyak plotting
plt.show()
```

```python
data = {'a': np.arange(50),#Link documentary: https://numpy.org/doc/stable/reference/generated/numpy.arange.html#numpy.arange
        'c': np.random.randint(0, 50, 50), #.randint(angka terkecil, angka terbesar, jumlah size). Link documentary: https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html
        'd': np.random.randn(50)} #Menghasilkan bilangan float dari standard normal distribution. Link documentary: https://numpy.org/doc/stable/reference/random/generated/numpy.random.randn.html
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100 #abs nilai absolut

data
```

```python
plt.scatter('a', 'b', data=data) #pola scatter plot dari key dictionary 'a' dan 'b'
plt.xlabel('entry a') #label x
plt.ylabel('entry b') #label y
plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
