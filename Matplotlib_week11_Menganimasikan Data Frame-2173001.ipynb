```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython import display
import numpy as np
```

```python
fig = plt.figure()#fungsi ini membuat suatu figure saja seperti kontainer besar
plt.xlim((0, 8))#Untuk menampilkan axes area gambaran plotting
plt.ylim((-4, 2))#xlim dan ylim untuk menentukan batas bawah dan atas dari sumbu x dan y secara implisit

print(plt.plot([]))#printout terkait objek plotting
```

```python
fig = plt.figure()
plt.xlim((0, 4))
plt.ylim((-2, 2))

line = plt.plot([])[0]#melakukan plotting dengan menyertakan data kosong menghasilkan objek matplotlib.Lines.Line2D untuk mengendalikan garisnya


def animate(i):#fungsi animate mempersiapkan nilai untuk X dan Y
    x = np.linspace(0, 4, 1000)#membuat deret bilangan mulai darin 0 sampai 4 sebanyak 1000 datapoint ditampung ke dalam variabel x
    y = np.sin(2 * np.pi * (x - 0.01 * i))#sekumpulan data kedua ditampung ke dalam variabel y. Menerapkan operasi sinus terhadap nilai yang terdapat pada variabel x
    line.set_data(x, y)#membentuk garis
    return line,#mengembalikan nilai line


anim = FuncAnimation(fig, animate, frames=200, interval=50)#FuncAnimation dibutuhkan untuk membuat animasi. Parameter: (figure, fungsi animasi, frames=jumlah frame untuk animasi, interval=kecepatan tiap frame ditampilkan)

video = anim.to_html5_video()#mengkonversi anim menjadi format html5 video.
html = display.HTML(video)#menampilkan video HTML dengan objek video ke variabel html
display.display(html)#menampilkan video HTML

plt.close()#plt.close agar tampilan statik dari frame animasi plotting terakhir tidak muncul
```
