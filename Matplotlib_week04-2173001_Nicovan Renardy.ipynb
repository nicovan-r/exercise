```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [25, 45, 55, 125, 225]#bentuk list
kategori = ['A', 'B', 'C', 'D', 'E']#bentuk list
```

```python
plt.bar(kategori, data)#plt.bar untuk menghasilkan bar plot, setidaknya butuh dua parameter.
plt.grid()#menampilkan grid

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
#Histogram untuk plotting data bertipe numerik dan bersifat continous. Bar plot untuk plotting yang bersifat categorical dan discrete
```

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, 
        data, 
        color='red', #mengatur warna dan alpha tranparansi bar
        alpha=0.25)#transparansi 25%

plt.grid(linestyle='--', #bentuk garis grid
         linewidth=2, #ketebalan garis grid
         axis='y', # pilihan: x, y, both untuk direction grid vertikal atau horizontal
         alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data1, label='Data 1')#label untuk penamaan data di legend
# plt.bar(kategori, data2, label='Data 2')
plt.bar(kategori, data2, label='Data 2', bottom=data1)#ada tambahan label karena terdiri dari dua data set, bottom=data1 penempatan data1 di bawah, kalau tidak ditentukan penempatannya maka barnya akan saling timpa tak tertumpuk

plt.legend()

plt.grid(linestyle='--', 
         linewidth=1, 
         axis='y',
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

```python
data1 = [25,85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']

x = np.arange(len(kategori))#fungsi np.arrange membentuk deret bilangan dari array ditentukan dr nilai di dalam (). len(kategori) menghasilkan panjang variable kategori
x
```

```python
#group bar plot untuk mengelompokan plotting barplot dari dua lebih data pada dataset yang berbeda dengan kategori yang sama, kedua dataset akan diplotting berdampingan satu dengan yang lainnya
width = 0.35#lebar bar dalam range 0.0 sampai 1.0

plt.bar(x-width/2, data1, width, label='Data 1')#param pertama berasosiasi dengan index dari kategori untuk menjadi acuan
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)#untuk menerapkan nama" sumbu x, disini berdasarkan kategori

plt.grid(linestyle='--', 
         linewidth=1, 
         axis='y',
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
plt.show()
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
plt.pie(data, #plt.pie memerlukan sekumpulan data bernilai integer
        labels=kategori,#labels=kategori bersifat optional
        autopct='%1.1f%%',#menampilkan presentase di setiap area
        startangle=90)#melakukan rotasi paada pyplot sesuai perputaran searah jarum jam saat nilainya bertambah besar

plt.title('Contoh Pie Plot')

plt.show()
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']#warna tiap kategori, bisa pakai kode komposisi warna RGB (diawali dengan simbol #) contoh: merah=e0, hijau=33, biru: 64
```

```python
plt.pie(data, 
        labels=kategori,
        colors=warna,
        autopct='%1.1f%%',
        startangle=90)

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']#warna tiap kategorinya
explode_var = [0., 0., 0.2, 0., 0.08]#bagiam mana yang mau kategorinya ditarik keluar
```

```python
plt.pie(data, 
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Contoh Exploded Pie Plot')

plt.show()
```

```python
data_uas=[
    ['Bejo', 70],#format penampung list
    ['Tejo', 83],
    ['Cecep', 62],
    ['Wati', 74],
    ['Karti', 71]
]
```

```python
data_uas
```

```python
#OO Style Table Plot
#table plot untuk menampilkan data dalam bentuk tabular bukan dalam format data frame, tetapi termasuk format figure dan axes.
fig, ax = plt.subplots()#menghasilkan 2 buah objek yang akan di tampung ke variable fig dan ax.

table = plt.table(cellText=data_uas, 
                  loc='center')#penempatan table plot

table.set_fontsize(14)#ukuran txt, ini merupakan atribut dari object table, jadi haru di tampung ke suatu atribut. Di sini ditampung ke atribut table
table.scale(1, 4) # ukuran kolom, baris

ax.axis(False)#untuk menyembunyikan/men-disable axes dari plotting yang dikakukan. Axes terdiri dari sumbu x dan y

plt.show()
```

```python
#Pyplot Style Table Plot
table = plt.table(cellText=data_uas, 
                  loc='center')

table.set_fontsize(14)
table.scale(1, 4) # ukuran kolom, baris
#dengan menggunaman pyplot style, figure dan axes tidak dideklarasikan secara eksplisit padahal diperlukan untuk mendisable axes, perlu memanggil objek akses
ax = plt.gca()#plt.gca untuk mendapatkan axes yang saat ini sedang aktif. gca singkatan dari get current axes mendapatkan referensi objek-objek yang sedang aktif di masukan ke variable ax.
ax.axis(False)

plt.show()
```
