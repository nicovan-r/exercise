```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0.0, 2.0, 0.01)#parameter np.arrange(starting point yang inklusif/disertakan, end point yang eksklusif/tidak disertakan, interval) untuk membuat deret
x
```

```python
s = np.sin(2 * np.pi * x) #menampung nilai operasi sinus dari dataset pada variabel x/nilai radian. s ini merupakan rumus trigonometri
s
```

```python
#OO Style plotting
fig, ax = plt.subplots() #menghasilkan dua buah nilai yang perlu ditampung ke dua variabel

ax.plot(x, s) #plotting x untuk sumbu x dan s untuk sumbu y
ax.set(xlabel='nilai x', 
       ylabel='nilai sin(x)',
       title='Visualisasi nilai sin')
ax.grid() #untuk menampilkan grid-grid pada plotting

plt.show() 
```

```python
#pyplot style plotting
s = np.sin(2 * np.pi * x) #menampung nilai operasi sinus dari dataset pada variabel x/nilai radian. s ini merupakan rumus trigonometri
s
plt.plot(x, s)
plt.xlabel('nilai x') #menyertakan x-label pada axes
plt.ylabel('nilai sin(x)') #menyertakan y-label pada axes
plt.title('Visualisasi nilai sin') #menyertakan title pada axes
```

```python
x1 = np.linspace(0.0, 5.0, 100)#perbedaan linspace dan arrange: linspace: menyertakan 3 parameter (nilai minimum inklusif, nilai max inklusif, jumlah data point)
x2 = np.linspace(0.0, 2.0, 100)

y1 = np.cos(2 * np.pi * x1) # cos cosinus
y2 = np.cos(2 * np.pi * x2)
```

```python
x1
```

```python
#OO Style multiple subplots
fig, (ax1, ax2) = plt.subplots(2, 1) #plt.subplots(jumlah baris, jumlah kolom). fig dan (ax1,ax2) merupakan tuple untuk menampung dua buah axes
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'ro-')#plt.plot(nilai x, nilai y, warna dan bentuk garis plotting). -- garis putus"
ax1.set_ylabel('Nilai $cos(x_1)$')#mengatur nama label. $cos(X_1)$ untuk menaruh latex style number X dan 1 kecil/subscript

ax2.plot(x2, y2, 'g-')
ax2.set_ylabel('Nilai $cos(x_2)$')#mengatur nama label. $cos(X_1)$ untuk menaruh latex style number X dan 1 kecil/subscript

ax2.set_xlabel('Nilai x')

plt.show()
```

```python
#Pyplot Style multiple subplots
plt.subplot(2, 1, 1)#parameter plt. subplot(jml baris, jml kolom, mengendalikan axes ke brpa)
plt.plot(x1, y1, 'ro-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai $cos(x_1)$')

plt.subplot(2, 1, 2)#parameter plt. subplot(jml baris, jml kolom, mengendalikan axes ke brpa)
plt.plot(x2, y2, 'g.-')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai $x$')

plt.show()
```

```python
mu, sigma = 100, 15 # mu untuk nilai mean dan sigma untuk nilai standard deviation

x = mu + sigma * np.random.randn(10000) # normal distribution untuk mengetahui sebaran data. nilai 10000 jumlah data point yang dihasilkan np.random.randn
x#keluar berupa struktur data array
```

```python
x.shape
```

```python
#Pyplot Style Histogram
plt.hist(x, #normal distribution
         bins=50, #bins untuk jumlah pengelompokan data berupa 50
         facecolor='g', #warna facecolor untuk diagram
         alpha=0.50) #nilai alpha untuk mengatur transparansi warna diagram

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45, 500, '$\mu=100,\ \sigma=15$')#text di dalam ploting. plt.text(kordinat peletakan teks x, kordinat peletakan teks y, string nama (untuk simbol kurung dengan backslash))
plt.grid()

plt.show()
```

```python
#OO Style Histogram
fig, ax = plt.subplots()

ax.hist(x, 
        bins=50, 
        facecolor='b', 
        alpha=0.25)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

ax.text(45, 500, '$\mu=100,\ \sigma=15$')
ax.grid()

plt.show()
```
