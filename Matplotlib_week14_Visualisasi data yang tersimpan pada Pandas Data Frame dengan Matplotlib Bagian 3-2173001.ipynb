```python
import matplotlib 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('./dataset/2020.csv')
#file csv ini akan dibentuk menjadi suatu objek data frame disimpan ke variabel df
df.head()
```

```python
#penggantian nama kolom data frame
df.rename(columns={
    'Country name': 'Country',
    'Perceptions of corruption': 'Corruption',
    'Freedom to make life choices': 'Freedom'
},
          inplace=True)
df.columns
```

```python
df[:5].plot(x='Country',
            y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
            kind='bar')#menampilkan isi data frame yang terdiri dari 5 baris pertama saja, kind='bentuk plot'


plt.title('Contoh Bar Plot')
plt.ylabel('Nilai')
plt.show()
```

```python
df[:5].plot(x='Country',
            y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
            kind='line')#sumbu x untuk Country, y untuk penilaian tertentu dari country. Menampilkan isi data frame yang terdiri dari 5 baris pertama saja

plt.title('Contoh Line Plot')
plt.ylabel('Nilai')
plt.show()
```

```python
df[:20].plot(y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
        kind='box')#mengambil nilai sebaran dari sumbu y untuk keseluruhan negara yang dimiliki. Semakin mengambil banyak data yang diambil maka sebaran datanya pun akan terpengaruh

plt.title('Contoh Box Plot')
plt.xlabel('Parameter')
plt.ylabel('Nilai')
plt.show()
```

```python
df[:5].plot(x='Corruption', y='Freedom', kind='scatter')#scatter plot digunakan untuk membandingkan keterkaitan antar dua variabel (sumbu x dan sumbu y).

plt.xlim((0,1))
plt.ylim((0,1))

plt.title('Contoh Scatter Plot')
plt.show()
```

```python
df.plot(x='Country', y='Corruption', kind='hist', bins=5)#mengindentifikasikan sebaran berbentu histogram dari keseluruhan negara. Parameter bins berkorelasi dengan jumlah kelas yang mau kita bentuk dari data yang dimiliki dan dikelompokkan. ketika bins dinaikan maka datanya menjadi lebih detail.

plt.title('Contoh Histogram')
plt.xlabel('Nilai')
plt.show()
```

```python
df.plot(x='Country',
        y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
        kind='hist',
        subplots=True,
        layout=(2, 2))

plt.suptitle('Contoh Subplots')#suptitile merepresentasikan super title ketika plotnya banyak
plt.tight_layout()#agar layoutnya rapi, tidak terjadi overlapping
plt.show()
```
